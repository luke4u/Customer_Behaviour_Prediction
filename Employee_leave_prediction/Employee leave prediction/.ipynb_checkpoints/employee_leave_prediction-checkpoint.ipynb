### 4. Data processing

```python
employee_df.head()
```

### 4.1 Separate numerical and categorical dataset

```python
X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
```

```python
X_cat.head()
```

### 4.2 Onehot encoding

```python
# onehotencoder = OneHotEncoder()
# X_cat = onehotencoder.fit_transform(X_cat).toarray()
X_cat = pd.get_dummies(X_cat)
```

```python
X_cat
```

```python
X_cat.head()
```

```python
X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome','Education', 
                           'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',
                           'JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate',
                           'NumCompaniesWorked','OverTime','PercentSalaryHike', 'PerformanceRating',
                           'RelationshipSatisfaction','StockOptionLevel',	'TotalWorkingYears',
                           'TrainingTimesLastYear', 'WorkLifeBalance','YearsAtCompany','YearsInCurrentRole', 
                           'YearsSinceLastPromotion','YearsWithCurrManager']]
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
```

```python
X_all
```

```python
y = employee_df['Attrition']
```

### 4.3 Scalling dataset

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
# 1/0
```

### 4.4 Train and test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

### 4.5 SMOTE oversampling

```python
oversampler = SMOTE(random_state=0)
X_smote_train, y_smote_train = oversampler.fit_sample(X_train,y_train)
```

### 5. Logistic regression model

```python
model = LogisticRegression()
model.fit(X_smote_train, y_smote_train)
```

```python
y_pred = model.predict(X_test)
```

#### by default, metrics is for class 1: leave

```python
print('Logistic regression model Accuracy {}%'.format(100 * accuracy_score(y_pred, y_test)))
print('Logistic regression model Precision {}%'.format(100 * precision_score(y_pred, y_test)))
print('Logistic regression model Recall {}%'.format(100 * recall_score(y_pred, y_test)))
print('Logistic regression model F1 {}%'.format(100 * f1_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_test, y_pred))
```

### 6. Random Forest model

```python
model = RandomForestClassifier()

model.fit(X_smote_train, y_smote_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('RF model Accuracy {}%'.format(100 * accuracy_score(y_pred, y_test)))
print('RF model Precision {}%'.format(100 * precision_score(y_pred, y_test)))
print('RF model Recall {}%'.format(100 * recall_score(y_pred, y_test)))
print('RF model F1 {}%'.format(100 * f1_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_test, y_pred))
```

### 7. ANN

```python
model = Sequential()
model.add(Dense(units = 50, activation = 'relu', input_shape = (50, )))
model.add(Dense(units = 500, activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(units = 500, activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(units = 50, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(X_smote_train, y_smote_train, epochs = 50, batch_size = 50)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['acc'])
plt.title('Model Accuracy Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training accuracy')
plt.legend(['Training accuracy'])
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```
